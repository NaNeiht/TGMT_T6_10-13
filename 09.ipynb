```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/AnImg/messiwc.JPG'
#Doc Anh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyen sang anh xam
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#Tao mot hinh ve moi
plt.figure()
plt.hist(im_array.flatten(), bins=255)
plt.title('Bieu do Histogram cua anh')
plt.xlabel('Gia tri cua pixel')
plt.ylabel('So Luong')
plt.show()
```

```python
# Lam noi bat duong bien
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
